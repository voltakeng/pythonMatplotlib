```python
import matplotlib.pyplot as plt 

room = ["A","B","C"]
boys = [10,15,5]
girls = [20,7,15]
```

```python
plt.bar(room,boys,label="Boys",color="blue")
plt.bar(room,girls,bottom=boys,label="Girls",color="orange")
plt.show()
```

```python
plt.stackplot(room,boys,girls,labels=["Boys","Girl"],colors=["green","pink"])
plt.show()
```
