```python
import matplotlib.pyplot as plt 

x = [1,2,3,4]
y = [1,2.5,3.5,3]
z = [1,1.5,1.5,3.5]
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("Mouth")
plt.ylabel("Sales")
plt.legend(["Mouse","Keyboard"],facecolor="pink")
plt.show()
```
