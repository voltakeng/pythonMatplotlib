```python
import matplotlib.pyplot as plt 

x = [1,2,3,4]
y = [1,2.5,3.5,3]
```

color support 
- b = blue        
- g = green      
- r = red  
- c = cyan        
- m = magenta
- y = yellow
- k = black
- w = white

```python
plt.bar(x,y,color='r')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
plt.bar(x,y,color=['r','b','y','m'],width=0.3,edgecolor="black",linewidth="5")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
plt.barh(x,y,color=['r','b','y','m'],edgecolor="black")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```
