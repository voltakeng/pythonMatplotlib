```python
import matplotlib.pyplot as plt 

x = [1,2,3,4]
y = [1,2.5,3.5,3]
```

marker support 
- '.' point
- 'o' cycle
- 'x' x marker
- '+' plus
- '*' star
- 's' square
- '|' vline
- '_' hline
- 'v' down
- '^' up
- '>' right
- '<' left

```python
plt.plot(x,y,color='m',marker='.')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```
