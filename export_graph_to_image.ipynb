```python
import matplotlib.pyplot as plt 

x = [1,2,3,4]
y = [1,2.5,3.5,3]
```

```python
#กราฟแท่ง
plt.bar(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")

plt.savefig("bar.jpg")

plt.show()
```

```python
#กราฟเส้น
plt.plot(x,y)
plt.xlabel("Mouth")
plt.ylabel("Sales")

plt.savefig("line.png",transparent=True)

plt.show()
```
