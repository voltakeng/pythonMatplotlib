```python
import matplotlib.pyplot as plt 

#กราฟวงกลม (เหมาะกับข้อมุลที่หาร100ลงตัว)
data = [15,30,45,60,40]
lang = ["PHP","Flutter","Nodejs","C#","Java"]
plt.pie(data,labels=lang)
plt.show()
```

```python
plt.pie(data,labels=lang,autopct="%.2f%%")
plt.show()
```

```python
plt.pie(data,labels=lang,autopct="%.2f%%")
plt.show()
```

```python
colors = ["purple","red","cyan","yellow","green"]
plt.pie(data,labels=lang,autopct="%.2f%%",colors=colors,shadow=True)
plt.show()
```

```python
colors = ["purple","red","cyan","yellow","green"]
plt.pie(data,labels=lang,autopct="%.2f%%",colors=colors,shadow=True,startangle=45)
plt.show()
```

```python
colors = ["purple","red","cyan","yellow","green"]
exp = [0,1,-0.5,0,0.2]
plt.pie(data,labels=lang,autopct="%.2f%%",colors=colors,shadow=True,explode=exp)
plt.show()
```
