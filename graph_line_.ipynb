```python
import matplotlib.pyplot as plt 

x = [1,2,3,4]
y = [1,2.5,3.5,3]
z = [1,1.5,1.5,3.5]
```

line support
- '-' solid
- '--' dashed
- '-.' dash-dot
- ':' dotted

```python
plt.plot(x,y,color='m',marker='>',linestyle='-.')
plt.plot(x,z)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```
